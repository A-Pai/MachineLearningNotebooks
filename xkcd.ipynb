```python
import matplotlib.pyplot as plt
import numpy as np

# Data to plot
dates = range(2010,2019)
percents = [0.01, 0.01, 0.05, 0.15, 0.35, 0.37, 0.40, 0.85, 0.99]

# Set the style to XKCD
plt.xkcd()

# Plot the percents
plt.plot(dates,percents, marker='o')

# Annotate and fine-tune
plt.title('Percentage Deep Learning at SMU', fontsize=13)
plt.annotate('Dr. Larson Hired', 
             xy=(dates[3], percents[3]), xytext=(dates[3]-0.2, percents[3]+0.4), 
             arrowprops=dict(facecolor='black', shrink=0.05))

plt.annotate('Python Machine Learning Offered', 
             xy=(dates[-3], percents[-3]), xytext=(dates[-3]+0.2, percents[-3]-0.2), 
             arrowprops=dict(facecolor='black', shrink=0.05))

plt.annotate('Like, Hope', 
             xy=(dates[-1], percents[-1]), xytext=(dates[-1]+3, percents[-1]-0.1), 
             arrowprops=dict(facecolor='black', shrink=0.05))

plt.xticks(dates,[str(x) for x in dates])

# Fine-tune the axis
plt.ylim([0.0, 1.0])

# Show and save
plt.show()
```
